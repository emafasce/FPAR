```
import matplotlib.pyplot as plt
import matplotlib

def plotting(loss_train, loss_val, acc_train, acc_val):
#Insert the list with the loss and accuracies of the training and validation phases and a title with a description of the nn used
  epochs=len(loss_train)
  x=[i+1 for i in range(epochs)]
  xdisp=[i for i in range(epochs) if i%50==0]
  font = {'family' : 'serif',
        'weight' : 'light',
        'size'   : 10}
  matplotlib.rc('font', **font)
  plt.figure()
  plt.plot(x, loss_train, 'b', label='Training loss', linewidth=0.8)
  plt.plot(x, loss_val, 'g', label='Validation loss', linewidth=0.8)
  plt.legend()
  plt.xlabel('Epochs')
  plt.xticks(xdisp)
  plt.ylim(ymin=0)
  plt.show()

  plt.figure()
  plt.plot(x, acc_train, 'b', label='Training accuracy', linewidth=0.8)
  plt.plot(x, acc_val, 'g', label='Validation accuracy',  linewidth=0.8)
  plt.legend()
  plt.xticks(xdisp)
  plt.xlabel('Epochs')
  plt.ylim(ymin=0)
  plt.show()
 
```
